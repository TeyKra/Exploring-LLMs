#### 5. Create a function to extract the text from your paragraphs.
#### 6. Ensure all your data is in lowercase.
#### 7. Remove tags, symbols, and numbers, and replace them with spaces.

```python
# Function to extract and clean paragraph texts
def extract_and_clean_paragraph_texts(soup):
    """
    Extracts the text from all paragraphs (<p>) in a BeautifulSoup object,
    removes HTML tags, unnecessary symbols, and numbers while preserving standard punctuation.

    Args:
        soup (BeautifulSoup): Parsed BeautifulSoup object.

    Returns:
        list: List containing the cleaned text of each paragraph.
    """
    paragraphs = soup.find_all('p')  # Find all <p> tags
    cleaned_paragraph_texts = []
    
    for para in paragraphs:
        text = para.get_text().lower()  # Convert to lowercase
        # Remove tags and numbers while keeping punctuation
        text = re.sub(r'[<>\[\]{}0-9]', ' ', text)  # Remove tags and numbers
        text = re.sub(r'[^\w\s.,;!?\'"]', ' ', text)  # Keep common punctuation
        text = re.sub(r'\s+', ' ', text).strip()  # Replace multiple spaces with a single one
        cleaned_paragraph_texts.append(text)
    
    return cleaned_paragraph_texts

# Call the function to extract paragraphs
paragraph_texts = extract_and_clean_paragraph_texts(soup)

# Display the extracted paragraphs
for i, text in enumerate(paragraph_texts, 1):
    print(f"Paragraph {i}:\n{text}\n")
```

#### 8. Tokenize.
##### Tokenize sentences:

```python
# Tokenize sentences from the cleaned paragraphs
all_sentences = []

for para in paragraph_texts:
    sentences = sent_tokenize(para, language='french')  # Specify French language for tokenization
    all_sentences.extend(sentences)

# Display the tokenized sentences
for i, sentence in enumerate(all_sentences, 1):
    print(f"Sentence {i}:\n{sentence}\n")
```

##### Tokenize words:

```python
# Tokenize words from the cleaned paragraphs
all_words = []

for para in paragraph_texts: 
    words = word_tokenize(para, language='french')  # Specify French language for tokenization
    all_words.extend(words)

# Display the tokenized words
for i, word in enumerate(all_words, 1):
    print(f"Word {i}: {word}")
```

#### 10. Define a unique function that processes your document by performing lemmatization, tokenization, conversion to lowercase, and punctuation removal using the previously defined functions.

```python
def process_document(soup):
    """
    Processes the document by performing lemmatization, tokenization, conversion to lowercase,
    and punctuation removal.

    Args:
        soup (BeautifulSoup): Parsed BeautifulSoup object.

    Returns:
        list: List of lemmatized and cleaned words.
    """
    # Initialize the French stemmer
    stemmer = SnowballStemmer("french")
    
    # Extract and clean paragraphs
    paragraphs = soup.find_all('p')  # Find all <p> tags
    cleaned_paragraph_texts = []
    
    for para in paragraphs:
        text = para.get_text().lower()  # Convert to lowercase
        # Remove tags and numbers while keeping punctuation
        text = re.sub(r'[<>\[\]{}0-9]', ' ', text)  # Remove tags and numbers
        text = re.sub(r'[^\w\s.,;!?\'"]', ' ', text)  # Keep common punctuation
        text = re.sub(r'\s+', ' ', text).strip()  # Replace multiple spaces with one
        cleaned_paragraph_texts.append(text)
    
    # Tokenize the words
    all_words = []
    for para in cleaned_paragraph_texts:
        words = word_tokenize(para, language='french')
        all_words.extend(words)
    
    # Remove punctuation
    all_words_no_punct = [word for word in all_words if word not in string.punctuation]
    
    # Lemmatize (stem) the words
    lemmatized_words = [stemmer.stem(word) for word in all_words_no_punct]
    
    return lemmatized_words

# Use the process_document function
processed_words = process_document(soup)

# Display the processed words
for i, word in enumerate(processed_words, 1):
    print(f"Processed word {i}: {word}")
```

#### 11. Create two dictionaries for possible inputs and responses

```python
# Dictionary of possible user greetings
GREETING_INPUTS = ["salut", "bonjour", "hey", "hello", "bonsoir", "coucou", "allo"]

# List of possible bot responses
GREETING_RESPONSES = ["Bonjour!", "Salut!", "Hey!", "Bonjour, comment puis-je vous aider?", "Salut, que puis-je faire pour vous?"]

def Welcome(user_response):
    """
    Prend en entrée la réponse de l'utilisateur et, si elle est dans les salutations,
    sélectionne une réponse aléatoire parmi les réponses possibles.

    Args:
        user_response (str): La réponse de l'utilisateur.

    Returns:
        str: Une réponse de salutation aléatoire si une correspondance est trouvée, sinon None.
    """
    # Convert the user's response to lowercase
    user_response = user_response.lower()

    # Check if the user's response matches a greeting
    if user_response in GREETING_INPUTS:
        # Select a random greeting response
        return random.choice(GREETING_RESPONSES)
    else:
        return None
```

#### 13. Create a generateResponse() function that takes user input, processes it, and generates a response if it exists in the database.

```python
def generateResponse(user_input):
    """
    Takes user input, processes it, and generates a response if it exists in the database.

    Args:
        user_input (str): The user's input.

    Returns:
        str: The chatbot's response.
    """
    # Add the user input to the list of sentences
    sentences = all_sentences.copy()
    sentences.append(user_input)

    # Create the TF-IDF vector
    vectorizer = TfidfVectorizer()
    tfidf = vectorizer.fit_transform(sentences)

    # Compute the cosine similarity between the user input and the sentences in the corpus
    cosine_similarities = cosine_similarity(tfidf[-1], tfidf[:-1])

    # Find the index of the sentence with the highest similarity
    idx = cosine_similarities.argsort()[0][-1]

    # Get the similarity score
    score = cosine_similarities[0][idx]

    # If the score is above a threshold, return the corresponding sentence
    if score > 0.1:
        response = sentences[idx]
    else:
        response = "Je suis désolé, je ne comprends pas votre question."

    return response
```

```python
print("Chatbot: Hello, I am an HR chatbot, how can I help you?")

while True:
    user_response = input("You: ")
    if user_response.lower() in ['goodbye', 'bye', 'quit', 'exit']:
        print("Chatbot: Goodbye! Have a great day.")
        break
    else:
        greeting = Welcome(user_response)
        if greeting is not None:
            print("Chatbot:", greeting)
        else:
            print("Chatbot:", generateResponse(user_response))
```
