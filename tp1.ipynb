#### b. Normalize your data: Create a function that:
- i. Removes punctuation
- ii. Removes ASCII and UTF-8 decoding artifacts
- iii. Removes unwanted symbols

```python
sent_tokens = nltk.sent_tokenize(data)

# Normalization function as defined earlier
def normalize_text(text):
    # 1. Remove punctuation
    text = ''.join(char for char in text if char not in string.punctuation)
    
    # 2. Remove non-ASCII characters
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    
    # 3. Remove unwanted symbols
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    return text

# Apply normalization to each sentence
normalized_sent_tokens = [normalize_text(sentence) for sentence in sent_tokens]

# Display some normalized sentences
print("The first 30 normalized sentences:\n")
for sentence in normalized_sent_tokens[:30]:
    print(sentence)
```

### 6- Define your lemmatizer and apply it to the cleaned data/tokens above (lemmatize())


```python
# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# List to store lemmatized sentences
lemmatized_sent_tokens = []

# Iterate through each normalized sentence
for sentence in normalized_sent_tokens:
    # Tokenize the sentence into words
    word_tokens = nltk.word_tokenize(sentence)
    
    # Lemmatize each word
    lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokens]
    
    # Reconstruct the sentence from lemmatized words
    lemmatized_sentence = ' '.join(lemmatized_words)
    lemmatized_sent_tokens.append(lemmatized_sentence)

# Display the first 5 lemmatized sentences for verification
print("The first 30 lemmatized sentences:\n")
for sentence in lemmatized_sent_tokens[:30]:
    print(sentence)
```

#### 7- Create 2 dictionaries: one for welcome inputs and one for welcome responses with possible inputs and responses

```python
# Dictionary of welcome inputs (greetings)
GREETING_INPUTS = [
    "hi", "hello", "hey", "good evening", "hi there", "greetings",
    "howdy", "yo", "hello there", "hiya", "good morning", "welcome",
    "what's up", "how are you", "how's it going", "what's new", "sup"
]

# Dictionary of welcome responses (greetings)
GREETING_RESPONSES = [
    "Hello! How can I assist you today?",
    "Hi! What can I do for you?",
    "Hey! Feel free to ask me your questions.",
    "Hello there! I'm here to help you.",
    "Hi! How can I be of service?",
    "Good day! How may I assist you?",
    "Great to see you! How can I help?",
    "Welcome! What can I do for you today?"
]
```

#### 8- Create a Welcome() function that takes the user's input. If the input is in the dictionary you created, it should randomly select a response from the provided responses.


```python
def Welcome(user_response):
    for word in user_response.lower().split():
        if word in GREETING_INPUTS:
            return random.choice(GREETING_RESPONSES)
    return None
```

#### 9- Create a generateResponse() function that responds to the user's question, processes it, and generates a response if one exists in the database.


```python
def generateResponse(user_response):
    robo_response = ''
    # Normalize the user's input
    user_response = normalize_text(user_response)
    # Tokenize the user's input
    user_response_tokens = nltk.word_tokenize(user_response)
    # Lemmatize the user's input
    user_response_lemmatized = [lemmatizer.lemmatize(word) for word in user_response_tokens]
    user_response_lemmatized = ' '.join(user_response_lemmatized)
    
    # Add the user's input to the lemmatized sentences
    lemmatized_sent_tokens.append(user_response_lemmatized)
    
    # TF-IDF Vectorization
    TfidfVec = TfidfVectorizer()
    tfidf = TfidfVec.fit_transform(lemmatized_sent_tokens)
    
    # Calculate cosine similarity
    vals = cosine_similarity(tfidf[-1], tfidf)
    idx = vals.argsort()[0][-2]
    flat = vals.flatten()
    flat.sort()
    req_tfidf = flat[-2]
    
    # Check similarity and generate a response
    if req_tfidf == 0:
        robo_response += "I'm sorry, I don't understand your question."
    else:
        robo_response += lemmatized_sent_tokens[idx]
    
    # Remove the user's input from the lemmatized sentences
    lemmatized_sent_tokens.pop()
    
    return robo_response
```

#### 10- Create the user interface that starts the chat with "Hello, I am an HR chatbot, how can I help you?" and responds to questions.

```python
print("Chatbot: Hello, I am an HR chatbot, how can I help you?")

while True:
    user_response = input("You: ")
    if user_response.lower() in ['goodbye', 'bye', 'quit', 'exit']:
        print("Chatbot: Goodbye! Have a great day.")
        break
    else:
        greeting = Welcome(user_response)
        if greeting is not None:
            print("Chatbot:", greeting)
        else:
            print("Chatbot:", generateResponse(user_response))
```
